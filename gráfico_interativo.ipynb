```python
import pandas as pd
```

```python
data= pd.read_csv('imigrantes_canada.csv')
data.set_index('País', inplace= True) # transformando a colona 'País' em um index
```

```python
anos= list(map(str, range(1980,2014)))
```

```python
Brasil_a= data.loc['Brasil',anos]
brasil_dict={'Anos':Brasil_a.index.tolist(),'Imigrantes':Brasil_a.values.tolist()}#Index.tolist() retorna uma lista dos valores.
Brasil= pd.DataFrame(brasil_dict)
Brasil.head()
```

###<font color= 'red' size= 8 font/> Alterando cores

<font color='yellow' size=6 fonnt/> Escolha de cores para diversas variáveis

```python
america_do_sul= data.query('Região == "América do Sul"').sort_values('Total', ascending= True)
america_do_sul.head()
```

```python
america_do_sul_c= america_do_sul.drop(['Continente','Região','Total'], axis=1) # 1= colunas, 2= linhas
A_sul= america_do_sul_c.T # O Ttransforma, colunas em linha e linhas em colunas
A_sul.head()
```

```python
cores= []
for pais in A_sul.columns:
    if pais == 'Brasil':
        cores.append('lawngreen') #Após instruções condicionais como if, elif ou else,
    elif pais == 'Argentina':       #você precisa identar o código que deve ser executado dentro desse bloco.
        cores.append('aqua')
    else:
        cores.append('lightgray')
```

```python
fig2= px.line(A_sul, x= A_sul.index, y=A_sul.columns,
             title= 'Imigrção América do sul ao Canadá \n (1980 à 2013)', color='País') # Escolhe uma cor para cada país
fig2.update_layout(width=1100, height=500,
             xaxis={'tickangle':-45},
             font_family='bold',
             font_size= 14,
             font_color='dark gray',
             title_font_color='black',
             title_font_size= 23,
             xaxis_title= 'Ano',
             yaxis_title='Número de Imigrantes')
fig2.update_traces(line_width=3)
```

```python
fig1= px.line(A_sul, x= A_sul.index, y=A_sul.columns,
             title= 'Imigrção América do sul ao Canadá \n (1980 à 2013)', color_discrete_sequence= cores) # Escolhe uma cor para cada país
fig1.update_layout(width=1100, height=500,
             xaxis={'tickangle':-45},
             font_family='bold',
             font_size= 14,
             font_color='dark gray',
             title_font_color='black',
             title_font_size= 23,
             xaxis_title= 'Ano',
             yaxis_title='Número de Imigrantes')
fig1.update_traces(line_width=3)
```

###<font size=8 font/> Salvando em HTML

```python
fig2.write_html('Imigrantes_Canada_AS.html')
```

```python
fig_motion.add_trace(
    go.Scatter(x=[Brasil['Anos'].iloc[0]], y=[Brasil['Imigrantes'].iloc[0]],
               mode='lines', name= 'Imigrantes', line_color='lawngreen', line= dict(width=3))
)
fig_motion.update_layout(title='<b> Imigração Brasil Canadá \n (1980 à 2013)</b>', title_x=0.14, # não existe o parametro text=
                  title_xanchor='left',font=dict(size=22),
                  xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),)
```

```python
fig_motion= go.Figure()
fig_motion.add_trace(
    go.Scatter(x=[Brasil['Anos'].iloc[0]], y=[Brasil['Imigrantes'].iloc[0]],
               mode='lines', name= 'Imigrantes', line= dict(width=3, color='lawngreen'))
)
fig_motion.update_layout(
  title=dict(
        text='<b>Imigração do Brasil para o Canadá no período de 1980 a 2013</b>', # text, é um paramêtro q deve as configurações do textos
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),

 updatemenus=[dict(
    type='buttons',    #define que o menu será composto por botões.
    showactive=False, #define que nenhum botão estará ativo inicialmente.
    buttons=[dict(label='Play',  #aqui nos escolhemos o label vulgo etiqueta do botão
                  method='animate', # aqui animamos os nossos dados
                  args=[None,     # o argumento None diz respeito se alguma camada do gráfico será afetada por uma animação
                        {'frame': {'duration': 100, 'redraw': True}, # O parâmetro frame define a duração de cada quadro da animação e a atualização de cada quadro.
                               'fromcurrent': True}])] #define se o quadro atual deve ser mantido ou se a animação deve ser iniciada do primeiro quadro.
)],
  width=1000,
  height=500
)
fig_motion.update_traces(showlegend=False, selector=dict(type='scatter')) # Ativa ou desativa a legenda


#A variável frames é uma lista de objetos Frame do Plotly, que contém as informações dos dados para cada quadro da animação.

frames = [go.Frame(data=[go.Scatter(x=Brasil['Anos'].iloc[:i+1], y=Brasil['Imigrantes'].iloc[:i+1], line=dict(color='lawngreen'))]) for i in range(len(Brasil))]
#O loop for é usado para criar um objeto Frame para cada ano no conjunto de dados, até o último ano.
#. A cada iteração, a lista frames é preenchida com um novo objeto Frame.
fig_motion.frames = frames
fig_motion.show()
```

<font size= 6 font/>Exemplo

```python
argentina_a= data.loc['Argentina',anos]
argentina_dict={'Anos':argentina_a.index.tolist(),'Imigrantes':argentina_a.values.tolist()}
Argentina= pd.DataFrame(argentina_dict)
Argentina['Anos']= Argentina['Anos'].astype(int) # primeiros convertemos a variável 'Anos' de uma strig para um número inteiro
Argentina.head()
```

```python

fig_motion2 = go.Figure()


fig_motion2.add_trace(
    go.Scatter(x=[Brasil['Anos'].iloc[0]], y=[Brasil['Imigrantes'].iloc[0]], mode='lines', name='Imigrantes do Brasil', line=dict(width=4, color='lawngreen'))
)


fig_motion2.add_trace(
    go.Scatter(x=[Argentina['Anos'].iloc[0]], y=[Argentina['Imigrantes'].iloc[0]], mode='lines', name='Imigrantes da Argentina', line=dict(width=4, color='aqua'))
)


fig_motion2.update_layout(
    title=dict(
        text='<b>Imigração do Brasil e da Argentina para o Canadá no período de 1980 a 2013',
        x=0.1,

        font=dict(size=18)
    ),
    xaxis=dict(range=[1980, 2013], autorange=False, title='<b>Ano</b>'),
    yaxis=dict(range=[0, 3000], autorange=False, title='<b>Número de imigrantes</b>'),
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 100, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1200,
    height=600
)


frames = []
for i in range(len(Brasil)):
    frame_data = [
        go.Scatter(x=Brasil['Anos'].iloc[:i+1], y=Brasil['Imigrantes'].iloc[:i+1], line= dict(color='lawngreen')),
        go.Scatter(x=Argentina['Anos'].iloc[:i+1], y=Argentina['Imigrantes'].iloc[:i+1], line= dict(color='aqua'))
    ]
    frame = go.Frame(data=frame_data)
    frames.append(frame)
fig_motion2.frames = frames

fig_motion2.show()
```
