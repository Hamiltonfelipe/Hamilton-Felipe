<font color='purple' size=15> Conhecendo os dados </font>

```python
import pandas as pd
```

```python
Dados= pd.read_csv('imigrantes_canada.csv')
Dados.head()
```

```python
Dados.info()
```

```python
Dados.set_index('País', inplace=True)
Dados.head()
```

```python
anos= list(map(str, range(1980,2014))) #A função map() executa uma função especificada para cada item em um iterável. O item é enviado para a função como um parâmetro.
anos[0:5] # Criamos uma lista onde todos os anos do nosso dataframe deixa de ser um 'int' pra se  tornar uma 'string', ou seja deixa de ser um número para se tornar um objeto
```

```python
Brasil=Dados.loc['Brasil', anos]
brasil_dict={'Ano': Brasil.index.tolist(),'Imigrantes': Brasil.values.tolist()}
Brasil=pd.DataFrame(brasil_dict)
Brasil.head()
```

<font color= 'purple' size=6> Processo de plotagem </font>
<hr>

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,8))#parametros são: larguram, altur(em polegadas)
plt.plot(Brasil['Ano'],Brasil['Imigrantes']) # x= eixo vertical (variável independente),y= eixo horizontal (variável dependente)
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, ])
plt.title('Immigração: Brasil Canadá')
plt.xlabel('Anos')
plt.ylabel('Número de Imigrantes')
plt.show()#Modulo de apresentação (interrompe execução do código)
```

### **EXEMPLO**

```python
Argentina=Dados.loc['Argentina', anos]
argentina_dict={'Ano': Argentina.index.tolist(),'Imigrantes': Argentina.values.tolist()}
Argentina=pd.DataFrame(argentina_dict)
```

```python
plt.figure(figsize=(15,8))
plt.plot(Argentina['Ano'],Argentina['Imigrantes'],color='aqua')
plt.plot(Brasil['Ano'],Brasil['Imigrantes'],color='lawngreen')

plt.legend(['Argentina','Brasil'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, ])
plt.title('Immigração: Argentina x Brasil (Canadá)')
plt.xlabel('Anos')
plt.ylabel('Número de Imigrantes')

plt.grid(True)
# aqui tiramos a conclussão de que a imigração de ambos os países se assemelhavam muito, até o ano de 2005 q houve uma subida exponencial do Brasil,
# logo após a argentina viver um pico de imigração e ter uma grande queda
```

```python
#é importante manter a mesma escala nos eixos X e Y em todos eles. Isso pode ser feito utilizando as funções set_xlim() e set_ylim().
fig,axs= plt.subplots(2, 2, figsize= (16,8))
fig.subplots_adjust(hspace= 0.5, wspace= 0.3) # hspace= espaçamento horizontal, wspace= espaçamento vertical
#o valor passado para esses parâmetros é um número decimal que representa a fração do tamanho da figura, por exemplo,
# 0.5 significa que o espaçamento será de 50% da altura/largura da figura.
axs[0,0].plot(Dados.loc['Brasil', anos], color= 'lawngreen')
axs[0,0].set_title('Brasil')

axs[0,1].plot(Dados.loc['Colômbia',anos], color= 'yellow')
axs[0,1].set_title('Colômbia')

axs[1,0].plot(Dados.loc['Argentina',anos], color= 'aqua')
axs[1,0].set_title('Argentina')

axs[1,1].plot(Dados.loc['Peru',anos], color= 'darkred')
axs[1,1].set_title('Peru')

for ax in axs.flat: #Usado para retornar um iterador 1-D sobre um determinado array de entrada usando o número de índice de cada um dos elementos do array.
   ax.xaxis.set_major_locator(plt.MultipleLocator(5))


for ax in axs.flat:
   ax.set_xlabel('Ano')
   ax.set_ylabel('Número de imigrantes')
   ax.grid(True, color= 'gray')

plt.show()
```

```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
meses= list(vendas_2022.keys())
meses
```

###<font size= 15> Comparação América do Sul

```python
America_S= Dados.query('Região == "América do Sul"').sort_values('Total', ascending= True) # para qualquer código q envolva string ele deve ser envolto por aspas
America_S.head()
```

```python
America_S.shape
```

```python
cores= []
for pais in America_S.index:
    if pais == 'Brasil':
        cores.append('lawngreen')
    else:
        cores.append('royalblue')

fig, ax= plt.subplots(figsize= (15,8))
ax.barh(America_S.index, America_S.Total,color=cores) # bar= barras verticais, barh= barras horizontais
ax.set_title('Imigração América do Sul ao Canadá (Total)\n1980 á 2013', fontsize=18, loc='left')
ax.set_xlabel('Número de imigrantes', fontsize= 13)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)
plt.show()
```

```python
import seaborn as sns
```

```python
sns.set_style('darkgrid') # escolhe o tema do seaborn no qual iremos trabalhar, usando o argmento set_theme, nos devemos especificar de forma correta o nome do tema
```

```python
top_10= Dados.sort_values('Total', ascending= False).head(10)# ascending= False, ordena do maior pro menor
top_10
```

```python
fig, ax= plt.subplots(figsize= (15,8))
ax= sns.barplot(data= top_10, x='Total', y=top_10.index, orient='h')
ax.set_title('Top 10 países imigrantes \n (Canadá - 1980 à 2013)',fontsize= 15, loc= 'left')
ax.set_xlabel('Número de imigrantes', fontsize= 12)
ax.set_ylabel('')

plt.show()
```

#<font size=6 font/> Utilizando paletas de cores do seaborn
<hr>

```python
def paleta(palette):
  fig, ax= plt.subplots(figsize= (15,7)) #def, utilizada para agrupar um conjunto de instruções em um bloco, permitindo que esse bloco seja executado quantas vezes forem necessárias.
  ax= sns.barplot(data= top_10, x='Total', y=top_10.index, orient='h', palette= palette)
  ax.set_title('Top 10 países imigrantes \n (Canadá - 1980 à 2013)',fontsize= 15, loc= 'left')
  ax.set_xlabel('Número de imigrantes', fontsize= 12)
  ax.set_ylabel('')

  plt.show()
  # Tudo isso está armazenado na função def: 'paleta'
```

<font color= 'purple' size= 5 font/> paleta de cores sequencias, são as paletas que apresentam uma variação gradual das cores

```python
#as_cmap é um argumento utilizado em funções do Seaborn para transformar um esquema de cores em um mapa de cores do Matplotlib.
#Quando as_cmap é definido como True, a função retorna um objeto de mapa de cores do Matplotlib.
#Esse argumento é útil para personalizar a visualização de cores em gráficos gerados com Seaborn.
sns.color_palette("crest", as_cmap=True)
```

https://seaborn.pydata.org/tutorial/color_palettes.html

```python
paleta('crest_r') # _r= reverse, ou seja, troca o sentido do degradê
```

<font color= 'red' size= 5 font/> paleta de cores divergentes, são as paletas que apresentam cores contrastantes

```python
sns.diverging_palette(300, 353, as_cmap=True)# O argumento 1= primeira, o 2= suas contrastante, 'center=dark'=o ponto médio é escuro em vez de claro
```

```python
paleta("rocket")
```

<font color= 'orange' size= 5 font/> paleta de cores categoricas, são as paletas utilizadas para demarcar diferentes categorias

```python
sns.color_palette("husl", 10) # argumento 1= ao sistema de cores, o 2= número de cores
```

```python
paleta('husl')
```
