# TMDB analise

```python
pd.read_csv('tmdb_5000_movies.csv')
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.original_language.unique()
```

```python
tmdb.original_language.value_counts().to_frame()
tmdb.original_language.value_counts().to_frame().reset_index ()
linguistica=tmdb.original_language.value_counts().to_frame().reset_index ()
```

```python
linguistica.columns=("lingua","recorrencia")
linguistica.head()
```

```python
sns.barplot(x="lingua", y="recorrencia", data = linguistica.head())
```

```python
pontuaçao_media.head()
```

# desmenbramendo de dados



```python
idioma=tmdb.original_language.value_counts()
idioma.head()
idioma_ingles=idioma.loc['en']
tudo=idioma.sum()
outros_idiomas = tudo - idioma_ingles
# objetivo demonstrar na imagem a maioria numerica dos filmes em lingua inglesa em relação ao geral
```

```python
dados = {
    "lingua": ('inglês', 'outros'),
    "total" : (idioma_ingles, outros_idiomas)
    }
```

```python
dados= pd.DataFrame(dados)
```

```python
sns.barplot(x="lingua", y="total",data=dados)
```

```python
nao_ingles= linguistica.query(" lingua != 'en'")
```

```python
print(nao_ingles)
```

```python
plt.figure(figsize=(16,4))
plt.title("numero de filmes em linguas diferentes do inglês")
sns.barplot(x= "lingua",y= "recorrencia",data= nao_ingles, palette= "rocket")
```
