# Manipulando os dados

## Criando colunas numéricas

```python
Dados=pd.read_csv('aluguel.csv', sep=';')
```

```python
Dados.fillna('0').inplace=True
```

```python
Dados['Valor_p_mes']=Dados['Condominio'] + Dados['Valor']
Dados.head()
```

```python
Dados['Valor_p_ano'] = Dados['Valor_p_mes'] *12 + Dados['IPTU']
Dados.head()
```

## Criando colunas categóricas

```python
Dados['Descricao'] = Dados['Tipo'] + ' em ' + Dados['Bairro'] + ' com ' + Dados['Quartos'].astype(str) + ' quarto(s) ' + ' e ' + Dados['Vagas'].astype(str) + ' vaga(s) '
Dados.head()
```

```python
Dados['Possui_suite']= Dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
Dados.head()
```

```python
Dados.to_csv('Dados_completos.csv', index= False)
```
